```python
import networkx as nx

def in_neighbors(graph, v):
    return set([i for i,j in graph.edges if j == v]+[v])

def out_neighbors(graph, v):
    return set([j for i,j in graph.edges if i == v]+[v])


def distributed_scc(graph):
    
    done = False
    
    while not done:
        # Identification on SCC

        n = graph.number_of_nodes()
        neigh_in = [in_neighbors(graph, node) for node in range(n)]
        neigh_out = [out_neighbors(graph, node) for node in range(n)]
        S = [set() for _ in range(n)]
        x = [set([i]) for i in range(n)]
        y = [1] * n
        z = [set() for _ in range(n)]
        r = [s.copy() for s in neigh_out]
        w = [False] * n

        while not all(w):
            print([x, y, z, r, w])
            x_prev, y_prev, z_prev, r_prev, w_prev = x.copy(), y.copy(), z.copy(), r.copy(), w.copy()
            for i in range(n):
                if not w[i]:
                    x[i] = set().union(*[x_prev[node] for node in neigh_in[i]])
                    y[i] = len(x[i])
                    z[i] = set(node for node in x[i] if y_prev[node] == y[i])
                    r[i] = set().union(r_prev[i],*(r_prev[node] for node in z[i]))
                    w[i] = y[i] == y_prev[i] and len(z[i]) == len(z_prev[i]) and len(r[i]) == len(r_prev[i])




        print([x, y, z, r, w])


        # Propagate SCC types
        def classify(i):
            
            if x[i] == z[i] and r[i] != z[i]:
                return "SOURCE"
            elif x[i] != z[i] and r[i] == z[i]:
                return "TARGET"
            elif x[i] == z[i] and r[i] == z[i]:
                return "SOURCE_TARGET
            else:
                return "MIXED"
        t = [classify(i) for i in range(n)]

        # Targets propose to sources

        targets = [{node for node in x[i] if t[node] == "SOURCE"} for i in range(n)]
        
        

        # Sources select and propose to target


        # Targets select sources
    
    
    
    return [(z[i], "SOURCE" if sorted(x[i]) == sorted(z[i]) and sorted(r[i]) != sorted(z[i]) else 
                   "TARGET" if sorted(x[i]) != sorted(z[i]) and sorted(r[i]) == sorted(z[i]) else 
                   "SOURCE_TARGET" if sorted(x[i]) == sorted(z[i]) and sorted(r[i]) == sorted(z[i]) else 
                   "MIXED") for i in range(n)]
```

```python
# Example usage:
# Create a graph using NetworkX
G = nx.DiGraph()
G.add_edges_from([(i-1,j-1) for i,j in [(3 , 4), (4 , 3), (3 , 2), (5 , 1) , (1 , 2), (2 , 1)]])

# Call the function with the graph
result = distributed_scc(G)
print(result)
```
